## Step 1: Download Audio from YouTube Playlist

Replace the `playlist_url` with your actual YouTube playlist link. The `yt-dlp` command will download each video as an MP3 file.

```python
# Import necessary libraries
import os
from pydub import AudioSegment

# Define the YouTube playlist URL
playlist_url = "https://music.youtube.com/playlist?list=PLbskaXy8C5WU" # <---- REPLACE WITH YOUR PLAYLIST LINK
output_dir = "downloaded_songs"

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Command to download audio from the playlist using yt-dlp
# -x: extract audio
# --audio-format mp3: convert audio to mp3
# --audio-quality 0: best audio quality
# --output: output template (%(title)s.%(ext)s will use the video title as filename)
command = f'yt-dlp -x --audio-format mp3 --audio-quality 0 --output "{output_dir}/%(title)s.%(ext)s" {playlist_url}'

print(f"Downloading audio from: {playlist_url}")
os.system(command)
print(f"All songs downloaded to: {output_dir}")
```

## Step 2: Add Cross-fade Between Songs

This code will iterate through the downloaded songs, load them, and then concatenate them with a cross-fade effect. You can adjust the `crossfade_duration_ms` as needed.

```python
from pydub import AudioSegment
from pydub.playback import play # Optional: for playing audio directly
import glob

# Path to the directory containing downloaded songs
songs_directory = "downloaded_songs"
output_filename = "mixed_playlist_with_crossfade.mp3"
crossfade_duration_ms = 5000 # 5 seconds crossfade

# Get all mp3 files from the directory, sorted alphabetically
song_files = sorted(glob.glob(os.path.join(songs_directory, "*.mp3")))

if not song_files:
    print(f"No MP3 files found in '{songs_directory}'. Please ensure songs were downloaded correctly.")
else:
    print(f"Found {len(song_files)} songs. Starting audio processing...")

    # Load the first song
    combined_audio = AudioSegment.from_mp3(song_files[0])

    # Iterate through the rest of the songs and add cross-fades
    for i in range(1, len(song_files)):
        print(f"Processing: {os.path.basename(song_files[i])}")
        next_song = AudioSegment.from_mp3(song_files[i])

        # Add a crossfade to the combined audio
        combined_audio = combined_audio.append(next_song, crossfade=crossfade_duration_ms)

    # Export the final combined audio
    combined_audio.export(output_filename, format="mp3")
    print(f"Successfully created '{output_filename}' with cross-fades!")

    # Optional: Play the combined audio (requires ffplay installed and on PATH)
    # print("Playing combined audio...")
    # play(combined_audio)
```
